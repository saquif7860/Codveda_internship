```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
df=pd.read_csv('iris_cleaned.csv')
df.head()
```

```python
sns.countplot(x='species', data=df)
plt.title("Count of Each Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
species_count = df['species'].value_counts()
plt.pie(species_count, labels=species_count.index, autopct='%1.1f%%', startangle=90)
plt.title("Species Proportion in Iris Dataset")
plt.axis('equal')
plt.show()
```

```python
plt.hist(df['petal_length'], bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()
```
